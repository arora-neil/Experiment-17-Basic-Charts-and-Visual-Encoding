```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days':['Mon','Tue','Wed','Thu','Fri'],
    'Study_Hours':[2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignments_Completed':[1,2,1,3,2]
}
df=pd.DataFrame(data)
df
```

```python
# 1. Line Chart (Trend)
plt.plot(df['Days'],df['Study_Hours'],marker='*')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

Other Common Marker Types:

'o' Circle

's' Square

'^' Triange


```python
# Advanced Line Graph
plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hours'],marker='o',color='blue',label='Study Hours')
plt.plot(df['Days'],df['Marks'],marker='s',color='red',label='Marks')
plt.title('Study Hours and Marks Over Days')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# 2. Bar Chart (Comparision)
plt.bar(df['Days'],df['Marks'],color='green')
plt.title('Marks Over Days')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Advanced Bar Graph
bars=plt.bar(df['Days'],df['Marks'],color='red')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title('Marks over Days')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y') # Adds horizontal grid lines
plt.show()
```

```python
# 3. Histogram (Distribution)
# Shows how values are spread
data = {
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=3,edgecolor='black',alpha=0.30)
plt.title("Marks Distribution")
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Example Data (more values for better histogram)
Marks=[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
# Calculate mean
mean_value=np.mean(Marks)
plt.figure(figsize=(7,4))
plt.hist(Marks,bins=6,color='skyblue',edgecolor='black')
# Mean line
plt.axvline(mean_value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title('Marks Distribution with Mean')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
# 4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
data = {
    'Days':['Mon','Tue','Wed','Thu','Fri'],
    'Study_Hours':[2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[6,7,6,8,7],
    'Assignments_Completed':[1,2,1,3,2]
}
df=pd.DataFrame(data)
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Conditional Scatter plot
# Example Category
df['Result']=['Fail','Pass','Fail','Pass','Pass']
colors=['red'if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title('Study Hours vs Marks (Result-Wise)')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Dataset - 2
import seaborn as sns
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics','Food',
                 'Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
df
```

```python
# Graphing using seaborn
# Line Plot
sns.lineplot(data=df,x='Day',y='Sales')
plt.title('Sales Over Days')
plt.show()
```

```python
# Bar plot
sns.barplot(data=df,x='Day',y='Profit')
plt.title('Profit by Days')
plt.show()
```

```python
# Histogram
sns.histplot(data=df,x='Sales',bins=5)
plt.title('Distribution of Sales')
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(data=df,x='Sales',y='Profit')
plt.title('Sales vs Profit')
plt.show()
```
